```python
wd = "../results-anonymized/experiment/"
end_of_game = pd.read_csv(wd+"end_of_game_effects_summary.csv", index_col=[0,1])

end_of_game['Effect Upper Error'] = end_of_game['Effect Upper Bound'] - end_of_game['Effect Size']
end_of_game['Effect Lower Error'] = end_of_game['Effect Size'] - end_of_game['Effect Lower Bound']



end_of_game
```

```python
plt.rcParams.update(rcParamsDefault)
plt.rcParams.update({
    "figure.dpi": 100,
    "figure.facecolor": "white", #"lightyellow",
    #"figure.facecolor": "lightyellow",
    
    "savefig.dpi": 150,
    
    "font.family": "san-serif",
    "font.sans-serif": "Helvetica",

    
    "text.color": "grey",
    "text.usetex": False,
    "hatch.color": "salmon",
    "hatch.linewidth": 5
})

def makesub(df, ax):
    conditions = ["('dodec', 'inter')", "('caveman', 'indep')", "('caveman', 'inter')"]
    
    for i, key in enumerate(conditions):
        alpha = 1# if df.at[key, "Effect P Value"] < .05 else 0.3
        ax.bar(x=i, 
               height=df.at[key, 'Effect Size'],
               #color = "r" if 'inter' in key else 'silver',
               color= "r" if key == "('dodec', 'inter')" else "silver" if key == "('caveman', 'indep')" else "silver",
               alpha=alpha,
               #hatch= "/" if key == "('caveman', 'inter')" else "",
               
               yerr=[[df.at[key, 'Effect Lower Error']], [df.at[key, 'Effect Upper Error']]],
               ecolor="k" if key == "('dodec', 'inter')" else "dimgrey", 
               error_kw=dict(lw=1, capsize=5, capthick=1, alpha=alpha),
              )
        if ((df.at[key, 'Effect Size'] > 0) & 
           (abs(df.at[key, 'Effect Lower Error']) > .5*abs(df.at[key, 'Effect Size']))):
            text_ypos = df.at[key, 'Effect Upper Bound']
            va='bottom'
        else:
            text_ypos = df.at[key, 'Effect Lower Bound']
            va='top'
            
        ax.text(x=i, 
                y=text_ypos,
                s="$\Delta$%.02f"%df.at[key, 'Effect Size'],
                va=va,
                ha='center',
                bbox={'pad':10, 'fc':'none', 'ec':'none'},
                color='dimgrey',
                alpha= 1 if alpha == 1 else 0.5
               )
        
    ax.text(x=2.4, 
            y=0,
            s="Baseline: %.02f"%df.at["('dodec', 'indep')", "Result Mean"],
            va="bottom" if df.at[key, 'Effect Size'] < 0 else "top",
            ha="right",
            fontsize=9
           )

    plt.hlines(0,-.5,2.5, color='grey', linewidth=1)
    plt.box(on=False)
    ax.set_xlim(-.5, 2.5)
    ax.set_xticks([])
    ax.set_yticks([])
    



plt.figure(figsize=(6.5, 7.5))
x0=.04
x1=.075
x2=.55
y1=.95

# similarity within camps, behavioral
ax1 = plt.subplot(3,2,1) 
df = end_of_game.swaplevel().loc['net spoke 95% similarity']
makesub(df,ax1)
#ax1.set_title('Behavioral Measure')
plt.figtext(x=x1, y=y1, va='top', color='k', size=10, 
            s='Behavioral Measure')
plt.figtext(x=x1+.22, y=y1, va='top', color='grey', size=10, 
            s='(22 elements)')


ax1.set_ylim(-.04, .195)
plt.figtext(x=x0, y=.96, s='Within-camp similarity\n($95^{th}$ percentile correlation)', rotation=90, va='top', ha='center')

# similarity within camps, self-report
ax2 = plt.subplot(3,2,2) 
df = end_of_game.swaplevel().loc['survey 95% similarity']
makesub(df,ax2)

plt.figtext(x=x2, y=y1, va='top', color='k', size=10, 
            s='Self-Report')
plt.figtext(x=x2+.125, y=y1, va='top', color='grey', size=10, 
            s='(11 elements)')
ax2.set_ylim(-.04, .195)

# similarity across camps, behavioral
ax3 = plt.subplot(3,2,3) 
df = end_of_game.swaplevel().loc['net spoke 5% similarity']
makesub(df,ax3)
ax3.set_ylim(-.135,.04)
plt.figtext(x=x0, y=.68, s='Across-camp similarity\n($5^{th}$ percentile correlation)', rotation=90, va='top', ha='center')

# similarity across camps, self-report
ax4 = plt.subplot(3,2,4) 
df = end_of_game.swaplevel().loc['survey 5% similarity']
makesub(df,ax4)
ax4.set_ylim(-.135,.04)

# PC1, behavioral
ax5 = plt.subplot(3,2,5) 
df = end_of_game.swaplevel().loc['net spoke PC1']
df.loc[:, df.columns != "Effect P Value"] *= 100  # scale to 100%
makesub(df,ax5)
ax5.set_ylim(-1, 17)
ax5.set_xticks([0,1,2])
ax5.set_xticklabels(["Effect of\nInterdependence", "Effect of\nPolarizing Network", "Effect of\nInterdependence +\nPolarizing Network"], 
                    rotation=90, color='grey')
ax5.tick_params(length=0)
plt.figtext(x=x0, y=.4, s='Alignment with "left-right" axis\n(% variance in $1^{st}$ component)', rotation=90, va='top', ha='center')

# PC1, self-report
ax6 = plt.subplot(3,2,6) 
df = end_of_game.swaplevel().loc['survey PC1']
df.loc[:, df.columns != "Effect P Value"] *= 100  # scale to 100%
makesub(df,ax6)
ax6.set_ylim(-1, 17)
ax6.set_xticks([0,1,2])
ax6.set_xticklabels(["Effect of\nInterdependence", "Effect of\nPolarizing Network", "Effect of\nInterdependence +\nPolarizing Network"], 
                    rotation=90, color='grey')
ax6.tick_params(length=0)

plt.suptitle('Effect of Interdependence and Network Structure', 
             x=0, y=1, 
             ha='left', va='top', fontsize=13, color='k')

plt.subplots_adjust(left=.08, right=.98, top=.92, bottom=.21, wspace=.17, hspace=.25)


plt.savefig('../documents/images/results.png', dpi=150)
plt.savefig('../documents/images/results_large.png', dpi=350)
```
