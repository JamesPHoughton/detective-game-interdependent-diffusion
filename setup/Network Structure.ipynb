```python
%pylab inline
import networkx as nx
```

```python
g1 = nx.dodecahedral_graph()
nx.draw_networkx(g1)
```

```python
# regular connected caveman
edgelist = [(0,1), (0,2), (0,3),
            (1,2), (1,3),
            (2,7),
            (4,5), (4,6), (4,7),
            (5,6), (5,7),
            (6,11),
            (8,9), (8,10), (8,11),
            (9,10), (9,11),
            (10,15),
            (12,13), (12,14), (12,15),
            (13,14), (13,15),
            (14,19),
            (16,17), (16,18), (16,19),
            (17,18), (17,19),
            (18,3)]

g4 = nx.from_edgelist(edgelist)
nx.draw_networkx(g4, alpha=.7)
```

```python
y = nx.degree_histogram(g1)
plt.bar(x=range(len(y)),height=y)
```

```python
y = nx.degree_histogram(g4)
plt.bar(x=range(len(y)),height=y)
```

```python
plt.hist([v for n1,d in nx.shortest_path_length(g1) for n2,v in d.items()], 
         bins=range(10), alpha=.5, label='Dodecahedron')
plt.hist([v for n1,d in nx.shortest_path_length(g4) for n2,v in d.items()], 
         bins=range(10), alpha=.5, label='Regular Connected Caveman')
plt.legend()
plt.xlabel('Length of Path Between Nodes')
plt.ylabel('Number of Pairs')
```
