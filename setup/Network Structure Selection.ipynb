# Network Structure Selection

This `Python 3` notebook explores various network structures with 20 nodes and fixed degree 3.

```python
g1 = nx.dodecahedral_graph()
nx.draw_networkx(g1)
g=g1
info.append({
    'name': 'Dodecahedron',
    'num_nodes': nx.number_of_nodes(g),
    'num_edges': nx.number_of_edges(g),
    'max_deg': max(nx.degree(g)),
    'min_deg': min(nx.degree(g)),
    'node_connectivity': nx.node_connectivity(g),
    'min_centrality': min(nx.betweenness_centrality(g).values()),
    'max_centrality': max(nx.betweenness_centrality(g).values()),
    'clustering': nx.average_clustering(g),
    'path_length': nx.average_shortest_path_length(g),
    'diameter': nx.diameter(g)
})
info[-1]
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D


# Create a figure and a 3D Axes
fig = plt.figure(figsize=(5,5))
ax = Axes3D(fig)

def init(G):
    pos = nx.spring_layout(G, dim=3)
    
    # Get number of nodes
    n = G.number_of_nodes()

    # Loop on the pos dictionary to extract the x,y,z coordinates of each node
    for key, value in pos.items():
        xi = value[0]
        yi = value[1]
        zi = value[2]

        # Scatter plot
        ax.scatter(xi, yi, zi, c='b', 
                   s=20+20*G.degree(key), 
                   edgecolors='b', 
                   alpha=0.9)

    # Loop on the list of edges to get the x,y,z, coordinates of the connected nodes
    # Those two points are the extrema of the line to be plotted
    for i,j in enumerate(G.edges()):
        x = np.array((pos[j[0]][0], pos[j[1]][0]))
        y = np.array((pos[j[0]][1], pos[j[1]][1]))
        z = np.array((pos[j[0]][2], pos[j[1]][2]))
        # Plot the connecting lines
        ax.plot(x, y, z, c='black', alpha=0.5)
        
    lim = .75
    ax.set_xlim(-1*lim, lim)
    ax.set_ylim(-1*lim, lim)
    ax.set_zlim(-1*lim, lim)
    ax.set_axis_off()
    
    return fig,

def animate(i):
    ax.view_init(elev=0., azim=i)
    return fig,

# Animate
anim = animation.FuncAnimation(fig, animate, init_func=lambda: init(g1),
                               frames=360, interval=20, blit=False)
# Save
anim.save('dodec_animation.gif', fps=30, writer='imagemagick')
```
